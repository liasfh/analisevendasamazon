## Análise de dados referente a tabela contendo informações dos produtos eletrônicos mais vendidos no Brasil pela site da Amazon.


```python
import pandas as pd

tabela = pd.read_csv("amz_br_produtos.csv")

display(tabela)
```

```python
# Analisar a versão resumida da tabela para verificar se tem alguma linha em branco ou se o tipo de linha esta relacionado corretamente
tabela.info()
```

```python
# tratando os dados, excluindo colunas desnecessárias
tabela = tabela.drop(["imgUrl","productURL"], axis=1)

display(tabela)
```

```python
# transformando o tipo de uma das linhas tabela em texto (str) para poder fazer filtragem de colunas.
tabela['isBestSeller'] = tabela['isBestSeller'].astype(str)
```

```python
tabela.info()
```

## 1 - Quantos pordutos são considerados Best Sellers?


```python
best_sellers = tabela['isBestSeller'].value_counts()

display(f"O número de produtos considerados mais vendidos pelo site da Amazon no Brasil é de: {best_sellers[1]}")
```

## 2 - Quais são os produtos considerados BestSeller?


```python
df_best_seller = tabela.loc[tabela['isBestSeller']=="True",:]
nome_produtos_bestseller = df_best_seller[["title"]]


display(nome_produtos_bestseller)

```

##  3 - Dos produtos Best Seller, quais foram os mais comprados e que mais faturaram no último mês segundo os dados?

```python
mais_comprados_mes = df_best_seller.loc[tabela['boughtInLastMonth']>=10000, :]
mais_comprados_mes = mais_comprados_mes.groupby("title").max(numeric_only=True)
mais_comprados_mes = mais_comprados_mes[["boughtInLastMonth"]]
mais_comprados_mes = mais_comprados_mes.sort_values("boughtInLastMonth", ascending=False)



display(mais_comprados_mes.head())
```

```python
# criando uma coluna nova chamada "faturamento" para nova análise.
comprados_ultimo_mes = df_best_seller.loc[tabela['boughtInLastMonth']>0, :]
comprados_ultimo_mes["Faturamento"] = comprados_ultimo_mes["boughtInLastMonth"] * comprados_ultimo_mes["price"]

#display(comprados_ultimo_mes)
```

```python
maior_faturamento = comprados_ultimo_mes[["title","Faturamento"]]
maior_faturamento = maior_faturamento.sort_values("Faturamento", ascending=False)

display(maior_faturamento.head())
```

## 4 - E qual foi o Best Seller que menos faturou em relação aos comprados no últimos mês?


```python
# Fazendo o tratamento dos dados e retirando as linhas cujo o preço é nulo.
df_best_seller2 = comprados_ultimo_mes.loc[comprados_ultimo_mes["price"]>0, :]



#display(df_best_seller2)
```

```python
menor_faturamento = df_best_seller2[["title", "Faturamento"]]
menor_faturamento = menor_faturamento.sort_values("Faturamento", ascending=True)

display(menor_faturamento.head())
```

## 5 - Dos produtos Best Sellers, quais deles receberam melhor avaliação considerando a coluna estrela?


```python
melhor_avaliacao = df_best_seller.loc[df_best_seller["stars"]>=4.8, :]
melhor_avaliacao = melhor_avaliacao.groupby("title").max()
melhor_avaliacao = melhor_avaliacao[["stars"]]
melhor_avaliacao = melhor_avaliacao.sort_values("stars", ascending=False)

display(melhor_avaliacao)
```

## 6 - Dos produtos não considerados Best Sellers, quais foram os mais vendidos no último mês segundos os dados?

```python
nao_best_seller = tabela.loc[tabela["isBestSeller"]== "False", :]
nao_best_seller = nao_best_seller.loc[nao_best_seller["boughtInLastMonth"]>0,:]
nao_best_seller = nao_best_seller[["title","boughtInLastMonth"]]
nao_best_seller = nao_best_seller.sort_values("boughtInLastMonth", ascending=False)

display(nao_best_seller.head())
```
